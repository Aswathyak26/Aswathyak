```python
# This program is for a bank account system. Users can deposit, withdraw, and check balance.
#This uses a dict to store account details and functions with conditions to perform operations.
#A new account is created if name is not in dict.
#Dict key is account holder name. Value is current balance.


info = {}
def balance_enquiry(name):
    if name not in info:
        print('New account created for ',name)
        info[name] = 0.0
    print('Current balance is: ',info[name])

def deposit(name,amt):
    if amt <= 0:
        print("Invalid amount")
        return
    info[name] = info.get(name,0.0)+amt
    print('Deposited ₹ ',amt)
    print('Updated balance:', info[name])

def withdraw(name,amt):
    if amt <= 0:
        print("Invalid amount")
        return
    balance=info.get(name,0.0)
    if amt<=balance:
        info[name]-=amt
        print('Withdrawn ₹ ',amt)
        print('Remaining balance:', info[name])
    else:
        print('Insufficient balance')


while True:
    holder=input('Enter account holder name (enter 1 to exit): ')
    if holder=='1':
        break
    if not holder.replace(' ', '').isalpha():
        print("Invalid name. Use only letters.")
        continue

    while True:
        print('How can I help you?')
        service=input(' Select a number from this (1.balance /2.deposit /3.withdraw /4.exit): ')
        if service not in ['1', '2', '3', '4']:
            print('Invalid input')
        elif service=='1':
            balance_enquiry(holder)
        elif service=='2':
            amt= input('Enter the amount to be deposited: ')
            if not amt.replace('.', '', 1).isdigit():
                print("Invalid input")
                continue
            amt = float(amt)
            deposit(holder,amt)
        elif service=='3':
            amt=input('Enter the amount to be withdrawn: ')
            if not amt.replace('.', '', 1).isdigit():
                print("Invalid input")
                continue
            amt = float(amt)
            withdraw(holder,amt)
        elif service=='4':
            print('Thankyou')
            break
        else:
            print('Service not available')

print('Full Account sheet: ')
print(info)
```
