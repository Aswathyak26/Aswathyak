```python
info = {}
def balance_enquiry(name):
  balance=info.get(name,0.0)
  print('Current balance is',balance)
  if name not in info:
     info[name] = 0.0

def deposit(name,amt):
  info[name] = info.get(name,0.0)+amt
  print('Deposited ₹ ',amt)
def withdraw(name,amt):
  balance=info.get(name,0.0)
  if amt<=balance:
   info[name]-=amt
   print('Withdrawn ₹ ',amt)
  else:
   print('Insufficient balance')

while True:
  holder=input('Enter account holder name (enter 1 to exit)')
  if holder=='1':
   break
  while True:
   service=input('How can I help you? select any number(1.balance/2.deposit/3.withdraw/4.exit)')
   if service=='1':
    balance_enquiry(holder)
   elif service=='2':
    amt=float(input('Enter the amount to be deposited'))
    deposit(holder,amt)
   elif service=='3':
    amt=float(input('Enter the amount to be withdrawn'))
    withdraw(holder,amt)
   elif service=='4':
    print('Thankyou')
    break
   else:
    print('Service not available')

print('Full Account sheet',info)
```

```python

def avg_value(data):
  x=sum(data)/len(data)
  return x

def weather(temp,hum):
  if hum>=75:
   return 'Rainy'
  if temp>=28 and hum<=60:
   return 'Sunny'
  if temp>=20 and hum>=70:
   return 'Cloudy'
  if temp<=20:
   return 'Cool'
  else:
   return 'Moderate'

temp=[]
hum=[]
region =input("Enter the name of region")
print("Enter the weather data for 7 days:")

for i in range(7):
  print('Day',i+1)
  day_temp = float(input("Enter temperature (°C): "))
  day_hum = float(input("Enter humidity (%): "))
  temp.append(day_temp)
  hum.append(day_hum)
  condition= weather(day_temp,day_hum)
  print('weather for day', i+1,'is',condition)



temp_avg= avg_value(temp)
hum_avg= avg_value(hum)

print('Average temperature is',temp_avg,'°C')
print('Average humidity is',hum_avg,'%')
```
