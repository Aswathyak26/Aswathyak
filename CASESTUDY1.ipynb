```python
# This program is for a bank account system. Users can deposit, withdraw, and check balance.
#This uses a dict to store account details and functions with conditions to perform operations.
#A new account is created if name is not in dict.
#Dict key is account holder name. Value is current balance.


info = {}
def balance_enquiry(name):
    if name not in info:
        print('New account created for ',name)
        info[name] = 0.0
    print('Current balance is: ',info[name])

def deposit(name,amt):
    if amt <= 0:
        print("Invalid amount")
        return
    info[name] = info.get(name,0.0)+amt
    print('Deposited ₹ ',amt)
    print('Updated balance:', info[name])

def withdraw(name,amt):
    if amt <= 0:
        print("Invalid amount")
        return
    balance=info.get(name,0.0)
    if amt<=balance:
        info[name]-=amt
        print('Withdrawn ₹ ',amt)
        print('Remaining balance:', info[name])
    else:
        print('Insufficient balance')


while True:
    holder=input('Enter account holder name (enter 1 to exit): ')
    if holder=='1':
        break
    if not holder.replace(' ', '').isalpha():
        print("Invalid name. Use only letters.")
        continue

    while True:
        print('How can I help you?')
        service=input(' Select a number from this (1.balance /2.deposit /3.withdraw /4.exit): ')
        if service not in ['1', '2', '3', '4']:
            print('Invalid input')
        elif service=='1':
            balance_enquiry(holder)
        elif service=='2':
            amt= input('Enter the amount to be deposited: ')
            if not amt.replace('.', '', 1).isdigit():
                print("Invalid input")
                continue
            amt = float(amt)
            deposit(holder,amt)
        elif service=='3':
            amt=input('Enter the amount to be withdrawn: ')
            if not amt.replace('.', '', 1).isdigit():
                print("Invalid input")
                continue
            amt = float(amt)
            withdraw(holder,amt)
        elif service=='4':
            print('Thankyou')
            break
        else:
            print('Service not available')

#print('Full Account sheet: ') #If we need to print full account details of all the holders.
#print(info)
```

```python
## This program takes input for 5 days and stores temp and hum values in lists.
#lists are used to calculate avg values and to determine weather conditions.
def avg_value(data):
    x=sum(data)/len(data)
    return x
def weather(temp,hum):
    if hum>=75:
        return 'Rainy'
    elif temp>=28 and hum<60:
        return 'Sunny'
    elif 20 <= temp < 28 and hum>=60:
        return 'Cloudy'
    elif temp<20:
        return 'Cool'
    else:
        return 'Moderate'
temp=[]
hum=[]
region =input("Enter the name of region")
print("Enter the weather data for 5 days:")
for i in range(5):
    print('Day',i+1)
    while True:
        t= input("Enter temperature (°C): ")
        h = input("Enter humidity (%): ")
        if not t.replace('.', '', 1).isdigit() or not h.replace('.', '', 1).isdigit():
            print("Invalid input. Please enter numbers.")
        else:
            break
    day_temp = float(t)
    day_hum = float(h)
    temp.append(day_temp)
    hum.append(day_hum)
    condition= weather(day_temp,day_hum)
    print('weather for day ', i+1,'is ',condition)
temp_avg = round(avg_value(temp),2)
hum_avg = round(avg_value(hum),2)
print('Average temperature is: ',temp_avg,'°C')
print('Average humidity is: ',hum_avg,'%')
```
